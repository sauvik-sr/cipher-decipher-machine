```python
from tkinter import *
from tkinter import messagebox
master=Tk()


###############################################################################################################
def vigAdDe(text,key,e):
    l=[]
    j=0
    for i in text:
        if(j==len(key)):
            j=0
        cValue=ord(key[j])-32
        oValue=ord(i)
        
        nValue=oValue-cValue
        if(nValue<32):
            nValue+=95
        l.append(chr(nValue))
        
        j+=1
        
    str1=''.join(map(str,l))   
    e.delete(0,END)
    e.insert(20,str1)
###############################################################################################################

def vigAdEn(text,key,e):
    l=[]
    j=0
    for i in text:
        if(j==len(key)):
            j=0
        cValue=ord(key[j])-32
        oValue=ord(i)
        
        nValue=oValue+cValue
        if(nValue>126):
            nValue-=95
        l.append(chr(nValue))
        
        j+=1
        
    str1=''.join(map(str,l))
    
    e.delete(0,END)
    e.insert(20,str1)

###############################################################################################################
def ceAdDe(text,key1,key2,e):
    l=[]
    for i in text:
        oValue=ord(i)
        if(key1>=95):
            key1=key1%95
        nValue=oValue-key1
        if(nValue<32):
            nValue+=126
            nValue-=31
        l.append(chr(nValue))
        key1+=key2
    str1=''.join(map(str,l))    
    e.delete(0,END)
    e.insert(20,str1)

###############################################################################################################

def ceAdEn(text,key1,key2,e):
    l=[]
    for i in text:
        oValue=ord(i)
        if(key1>=95):
            key1=key1%95
        nValue=oValue+key1
        if(nValue>126):
            nValue-=126
            nValue+=31
        l.append(chr(nValue))
        key1+=key2
    str1=''.join(map(str,l))    
    e.delete(0,END)
    e.insert(20,str1)

###############################################################################################################

def railDe(text,key,e):
    
    n=(key)*2
    l = [['0' for i in range(len(text))]for j in range(key)]
    length=0
    
    for i in range(key):
        
        if(i!=key-1):
            n=n-2
        else:
            n=(key-1)*2
            
        j=i
        while(j<len(text)):
            l[i][j]=text[length]
            length+=1
            j=j+n
            
    l1=[]
    j=0
    flag=0
    for i in range(len(text)):
        l1.append(l[j][i])
        
        if(flag==0):
            j+=1
        else:
            j-=1
            
        if(j==key):
            flag=1
            j=key-2
        elif(j==-1):
            flag=0
            j=0
            
    str1=''.join(map(str, l1))
    e.delete(0,END)
    e.insert(20,str1)

###############################################################################################################

def railEn(text,key,e):
    
    n=(key)*2
    l=[]
    
    for i in range(key):
        l1=[]
        a=i
        
        if(i!=key-1):
            n=n-2
        else:
            n=(key-1)*2
            
        while(a<len(text)):
            l1.append(text[a])
            a=a+n
        
        l.append(l1)
    
    l1=[]
    for i in range(key):
        str2 = ''.join(map(str, l[i]))
        l1.append(str2)
        
    str1 = ''.join(map(str, l1))
    e.delete(0,END)
    e.insert(20,str1)
    
###############################################################################################################

def vigCon(text,key,x,e):
    
    t=text.get()
    k=key.get()
    
    flag=0
    
    for i in t:
        if(ord(i)>126 or ord(i)<32):
            flag=1
            break
            
    for i in k:
        if(ord(i)>126 or ord(i)<32):
            flag=1
            break
    
    if(flag==1 or len(t)==0 or len(k)==0):
        messagebox.showwarning("warning","Please follow the rules")
    else:
        if(x==0):
            vigAdEn(t,k,e)
        else:
            vigAdDe(t,k,e)
    
###############################################################################################################

def ceaCon(text,key1,key2,x,e):
    
    t=text.get()
    k1=key1.get()
    k2=key2.get()
    
    if(k1<0 or k2<0 or len(t)==0):
        messagebox.showwarning("warning","Please follow the rules")
    else:
        if(x==0):
            ceAdEn(t,k1,k2,e)
        else:
            ceAdDe(t,k1,k2,e)

###############################################################################################################

def railCon(Text,key,x,e):
    
    t=Text.get()
    k=key.get()
    
    if(len(t)<=k or len(t)==0):
        messagebox.showwarning("warning","Please follow the rules")
    else:
        if(x==0):
            railEn(t,k,e)
        else:
            railDe(t,k,e)
            
###############################################################################################################            
def vigDeGui():
    
    thirdWin=Toplevel(master)
    thirdWin.title("Vigenère Cipher Decryption")
    thirdWin.geometry("600x350")
    
    cText=StringVar()
    key=StringVar()
    
    F=0
    
    railLabel=Label(thirdWin,text="Enter/Paste the cipher text and key",font=('Arial',15)).grid(row=0,column=1)
    railLabel2=Label(thirdWin,text="Note:").grid(row=2,column=1, sticky=W)
    railLabel3=Label(thirdWin,text="1. Key is a string value and can't be empty").grid(row=3,column=1, sticky=W)
    railLabel4=Label(thirdWin,text="2. Plain Text allows all special charachters and english letters").grid(row=4,column=1, sticky=W)
    railLabel5=Label(thirdWin,text="3. Plain text cannot be empty").grid(row=5,column=1, sticky=W)
    
    Label(thirdWin,text ="Cipher Text").grid(row=7,column=1,pady=20)
    Label(thirdWin,text ="Key").grid(row=8,column=1)

    e1=Entry(thirdWin,textvariable=cText).grid(row=7,column=2,pady=10)
    e2=Entry(thirdWin,textvariable=key).grid(row=8,column=2,pady=10)
    
    Label(thirdWin,text ="Plain Text").grid(row=11,column=1)
    e=Entry(thirdWin)
    e.grid(row=11,column=2)
    bt1=Button(thirdWin, text="Submit", command=lambda:vigCon(cText,key,1,e)).grid(row=12,column=2,pady=20)
    
###############################################################################################################            
def vigEnGui():
    
    thirdWin=Toplevel(master)
    thirdWin.title("Vigenère Cipher Encryption")
    thirdWin.geometry("600x350")
    
    pText=StringVar()
    key=StringVar()
    
    railLabel=Label(thirdWin,text="Enter/Paste the plain text and key",font=('Arial',15)).grid(row=0,column=1)
    railLabel2=Label(thirdWin,text="Note:").grid(row=2,column=1, sticky=W)
    railLabel3=Label(thirdWin,text="1. Key is a string value and can't be empty").grid(row=3,column=1, sticky=W)
    railLabel4=Label(thirdWin,text="2. Plain Text allows all special charachters and english letters").grid(row=4,column=1, sticky=W)
    railLabel5=Label(thirdWin,text="3. Plain text cannot be empty").grid(row=5,column=1, sticky=W)
    
    Label(thirdWin,text ="Plain Text").grid(row=7,column=1,pady=20)
    Label(thirdWin,text ="Key").grid(row=8,column=1)
    
    e1=Entry(thirdWin,textvariable=pText).grid(row=7,column=2,pady=10)
    e2=Entry(thirdWin,textvariable=key).grid(row=8,column=2,pady=10)
    
    Label(thirdWin,text ="Cipher Text").grid(row=11,column=1)
    e=Entry(thirdWin)
    e.grid(row=11,column=2)
    bt1=Button(thirdWin, text="Submit", command=lambda:vigCon(pText,key,0,e)).grid(row=12,column=2,pady=20)
    
            
###############################################################################################################

def ceaDeGui():
    
    thirdWin=Toplevel(master)
    thirdWin.title("Ceaser Cipher Decryption")
    thirdWin.geometry("600x390")
    
    cText=StringVar()
    key1=IntVar()
    key2=IntVar()
    
    railLabel=Label(thirdWin,text="Enter/Paste the cipher text and key",font=('Arial',15)).grid(row=0,column=1)
    railLabel2=Label(thirdWin,text="Note:").grid(row=2,column=1, sticky=W)
    railLabel3=Label(thirdWin,text="2. Both Key should be greater than 0").grid(row=4,column=1, sticky=W)
    railLabel4=Label(thirdWin,text="1. Both key will have integer value").grid(row=3,column=1, sticky=W)
    railLabel5=Label(thirdWin,text="3. Plain text cannot be empty").grid(row=5,column=1, sticky=W)
    
    Label(thirdWin,text ="Cipher Text").grid(row=7,column=1,pady=20)
    Label(thirdWin,text ="Key 1").grid(row=8,column=1)
    Label(thirdWin,text ="Key 2").grid(row=9,column=1)

    e1=Entry(thirdWin,textvariable=cText).grid(row=7,column=2,pady=10)
    e2=Entry(thirdWin,textvariable=key1).grid(row=8,column=2,pady=10)
    e3=Entry(thirdWin,textvariable=key2).grid(row=9,column=2,pady=10)
    
    Label(thirdWin,text ="Plain Text").grid(row=11,column=1)
    e=Entry(thirdWin)
    e.grid(row=11,column=2)
    bt1=Button(thirdWin, text="Submit", command=lambda:ceaCon(cText,key1,key2,1,e)).grid(row=12,column=2,pady=20)
            
###############################################################################################################
def ceaEnGui():
    
    thirdWin=Toplevel(master)
    thirdWin.title("Ceaser Cipher Encryption")
    thirdWin.geometry("600x390")
    
    pText=StringVar()
    key1=IntVar()
    key2=IntVar()
    
    railLabel=Label(thirdWin,text="Enter/Paste the plain text and key",font=('Arial',15)).grid(row=0,column=1)
    railLabel2=Label(thirdWin,text="Note:").grid(row=2,column=1, sticky=W)
    railLabel3=Label(thirdWin,text="2. Both Key should be greater than 0").grid(row=4,column=1, sticky=W)
    railLabel4=Label(thirdWin,text="1. Both key will have integer value").grid(row=3,column=1, sticky=W)
    railLabel5=Label(thirdWin,text="3. Plain text cannot be empty").grid(row=5,column=1, sticky=W)
    
    Label(thirdWin,text ="Plain Text").grid(row=7,column=1,pady=20)
    Label(thirdWin,text ="Key 1").grid(row=8,column=1)
    Label(thirdWin,text ="Key 2").grid(row=9,column=1)

    e1=Entry(thirdWin,textvariable=pText).grid(row=7,column=2,pady=10)
    e2=Entry(thirdWin,textvariable=key1).grid(row=8,column=2,pady=10)
    e3=Entry(thirdWin,textvariable=key2).grid(row=9,column=2,pady=10)
    
    Label(thirdWin,text ="Chiper Text").grid(row=11,column=1)
    e=Entry(thirdWin)
    e.grid(row=11,column=2)
    bt1=Button(thirdWin, text="Submit", command=lambda:ceaCon(pText,key1,key2,0,e)).grid(row=12,column=2,pady=20)
            
###############################################################################################################
def railDeGui():
    
    thirdWin=Toplevel(master)
    thirdWin.title("Rail Fence Cipher Decryption")
    thirdWin.geometry("600x350")
    
    cText=StringVar()
    key=IntVar()
    
    railLabel=Label(thirdWin,text="Enter/Paste the cipher text and key",font=('Arial',15)).grid(row=0,column=1)
    railLabel2=Label(thirdWin,text="Note:").grid(row=2,column=1, sticky=W)
    railLabel3=Label(thirdWin,text="1. Key should be integer value").grid(row=3,column=1, sticky=W)
    railLabel4=Label(thirdWin,text="2. Length of text should be greater than 1").grid(row=4,column=1, sticky=W)
    railLabel5=Label(thirdWin,text="3. Length of text should be greater or equal to key").grid(row=5,column=1, sticky=W)
    railLabel6=Label(thirdWin,text="4. Plain text cannot be empty").grid(row=6,column=1, sticky=W)
    
    Label(thirdWin,text ="Cipher Text").grid(row=7,column=1,pady=20)
    Label(thirdWin,text ="Key").grid(row=8,column=1)

    e1=Entry(thirdWin,textvariable=cText).grid(row=7,column=2,pady=10)
    e2=Entry(thirdWin,textvariable=key).grid(row=8,column=2,pady=10)
    
    Label(thirdWin,text ="Plain Text").grid(row=11,column=1)
    e=Entry(thirdWin)
    e.grid(row=11,column=2)
    bt1=Button(thirdWin, text="Submit", command=lambda:railCon(cText,key,1,e)).grid(row=12,column=2,pady=20)

###############################################################################################################
def railEnGui():
    
    thirdWin=Toplevel(master)
    thirdWin.title("Rail Fence Cipher Encryption")
    thirdWin.geometry("600x350")
    
    pText=StringVar()
    key=IntVar()
    
    railLabel=Label(thirdWin,text="Enter/Paste the plain text and key",font=('Arial',15)).grid(row=0,column=1)
    railLabel2=Label(thirdWin,text="Note:").grid(row=2,column=1, sticky=W)
    railLabel3=Label(thirdWin,text="1. Key should be integer value").grid(row=3,column=1, sticky=W)
    railLabel4=Label(thirdWin,text="2. Length of text should be greater than 1").grid(row=4,column=1, sticky=W)
    railLabel5=Label(thirdWin,text="3. Length of text should be greater or equal to key").grid(row=5,column=1, sticky=W)
    railLabel6=Label(thirdWin,text="4. Plain text cannot be empty").grid(row=6,column=1, sticky=W)
    
    Label(thirdWin,text ="Plain Text").grid(row=7,column=1,pady=20)
    Label(thirdWin,text ="Key").grid(row=8,column=1)

    e1=Entry(thirdWin,textvariable=pText).grid(row=7,column=2,pady=10)
    e2=Entry(thirdWin,textvariable=key).grid(row=8,column=2,pady=10)
    
    Label(thirdWin,text ="Chiper Text").grid(row=11,column=1)
    e=Entry(thirdWin)
    e.grid(row=11,column=2)
    bt1=Button(thirdWin, text="Submit", command=lambda:railCon(pText,key,0,e)).grid(row=12,column=2,pady=20)
    
###############################################################################################################

def vigenèreCipher():
    
    secondWin=Toplevel(master)
    secondWin.title("Vigenère Cipher")
    secondWin.geometry("300x300")
    
    railLabel=Label(secondWin,text="Choose one of the operation",font=('Arial',15)).pack(padx=20,pady=25)
    bt1=Button(secondWin,text="Encrypt",font=('Arial',15), command=vigEnGui).pack(padx=20,pady=20)
    bt1=Button(secondWin,text="Decrypt",font=('Arial',15), command=vigDeGui).pack(padx=20,pady=20)

###############################################################################################################

def ceaserCipher():
    
    secondWin=Toplevel(master)
    secondWin.title("Ceaser Cipher Advance")
    secondWin.geometry("300x300")
    
    railLabel=Label(secondWin,text="Choose one of the operation",font=('Arial',15)).pack(padx=20,pady=25)
    bt1=Button(secondWin,text="Encrypt",font=('Arial',15), command=ceaEnGui).pack(padx=20,pady=20)
    bt1=Button(secondWin,text="Decrypt",font=('Arial',15), command=ceaDeGui).pack(padx=20,pady=20)

###############################################################################################################

def railCipher():
    
    secondWin=Toplevel(master)
    secondWin.title("Rail Fence Cipher")
    secondWin.geometry("300x300")
    
    railLabel=Label(secondWin,text="Choose one of the operation",font=('Arial',15)).pack(padx=20,pady=25)
    bt1=Button(secondWin,text="Encrypt",font=('Arial',15), command=railEnGui).pack(padx=20,pady=20)
    bt1=Button(secondWin,text="Decrypt",font=('Arial',15), command=railDeGui).pack(padx=20,pady=20)
    
##############################################################################################################

master.title("Main Page")

main_label=Label(master, text="Welcome To Encription/Decription Machine",font=('Arial',15)).pack(padx=20,pady=20)
second_label=Label(master, text="Choose One of the Cipher",font=('Arial',10)).pack(padx=20,pady=25)

mw_bt1=Button(master, text="Rail Fence cipher",font=('Arial',15), command=railCipher).pack(padx=20,pady=20)
mw_bt2=Button(master, text="Ceaser Cipher Advance",font=('Arial',15), command=ceaserCipher).pack(padx=20,pady=20)
mw_bt3=Button(master, text="Vigenère Cipher",font=('Arial',15), command=vigenèreCipher).pack(padx=20,pady=20)
master.geometry("500x500")
mainloop()
```
